### 4.2 Add column names through list of series

### 4.3 Referencing and indexing a DataFrame

#### 4.3.1 Referencing columns

```python
# Referencing a column by column label (in this case, "studyName")
penguin["studyName"]
```

```python
# Referencing multiple columns by a list of column labels
# (in this case, the columns "studyName" and "Species")
penguin[["studyName", "Species"]]
```

```python
# refer to a single column in the data frame
print(penguin["Culmen_Length"])

# what is the pandas type and native type of the column you just printed (not of the contents)
print(type(penguin["Culmen_Length"]))
penguin["Culmen_Length"].dtype
```

The fastest way to reference rows and columns in pandas is by using loc and iloc. Just be aware that while sometimes they are interchangeable, other times they are not. iloc is going to want integers as the inputs, whereas loc is looking for label names (which can be integers, but can be other datatypes as well).

Refer to [this article](https://www.google.com/url?q=https%3A%2F%2Ftowardsdatascience.com%2Fhow-to-use-loc-and-iloc-for-selecting-data-in-pandas-bd09cb4c3d79) for a longer explanation of the similarities and differences between loc and iloc.

#### 4.3.2. Referencing rows
* **loc** gets rows (and/or columns) with particular labels.

* **iloc** gets rows (and/or columns) at integer locations.



```python
# print few rows for reference
penguin.head()
```

```python
# Access first row using zero based index
# in this case 10th row
penguin.iloc[0]
```

```python
# Reference multiple rows by index label (in this case the index label 10 through 12
# Returns a DataFrame , end index is not inclusive
penguin.iloc[10:12]
```

```python
# Access first row by its label
# HINT: look at column to the extreme left
# In this case the row is row 0

penguin.loc[0]
```

```python
# Reference multiple rows by row name (in this case rows 10 through 12)
# Note that this time the range includes the stop number
penguin.loc[10:12]
```

#### 4.3.3 Referencing rows and columns at the same time

```python
# Referencing a subset of rows and columns using index and column labels
# Note that we're using a range of column labels instead of a list
# Make sure that your column range starts with the leftmost label
# in this case start from first five rows continaing columns of Species until Stage
print(penguin.loc[0:4, "Species":"Stage"])

print(penguin.iloc[0:4, 0:4])
```

## 7.Apply method

The apply method takes as input a function, which can be both user defined or built in aggreagate functions (e.g. min, max, sum, etc.).

By default this is applied to every column (i.e. series) if column name is not specified in the function call, but can be changed by using the argument axis = "columns"

```python
# user defined function
def my_format(x):
  return f"{x:.2f}"

#format floats
penguin["Culmen_Length"] = penguin["Culmen_Length"].apply(my_format)
penguin.head()

```

## 8.Filtering rows
We can filter rows of a DataFrame using conditional statements to test if values within one or more columns meet the provided criteria. This is helpful if say we want to remove unnecessary rows of data or observe a specific range of data.

```python
# filter all female penguin species and the regions and Islands they come from
# uses the logical and comparison operators
penguin[penguin["Sex"] == "FEMALE"][["Species", "Region", "Island"]]
```

```python
# filter all rows with Adelie Penguin (Pygoscelis adeliae) Species that have body mass less than the mean body mass
penguin[(penguin["Species"] == "Adelie Penguin (Pygoscelis adeliae)") \
        & (penguin["Body Mass"] < penguin["Body Mass"].mean())]
```

## 9.Missing data

 - **Remove the data points**: proportion is small, the missingness is random (Consult your statistician! Compare mean/ mode/ median to decide )
 - **Replace with known default values**: domain defaults or random number


```python
# check for missing values - identify columns that would be interesting to look at
# identify columns that have missing information
penguin.info()
```

```python
# check for particular columns with missing values
# in this case comments column

penguin["Comments"].isna().sum()
```

```python
# check for the number of columns with missing values
# in this case Culmen_Depth
penguin["Culmen_Depth"].isna().sum()
penguin["Flipper_Length"].isna().sum()
```

To clean the data you have four options:
- remove the entire column (if it is not useful for our explorations)
- replace missing values with known defualt
- replace missing values with very small or large values
- remove rows that have the missing data

```python
# we decide to drop the comments column after looking at the column (more than 50% of data is missing)
# remove Comments column

penguin.drop(columns = "Comments", inplace = True)
penguin.info()

# you dont have to drop it, in case where yo want to keep the column - use fillna to add a string with known default value.
```

```python
# For some columns because they are crucial for helping us determine species features
# Replace values in culmen_depth and Flipper_length with averages
penguin["Culmen_Depth"].fillna(penguin["Culmen_Depth"].mean(), inplace = True)
penguin["Flipper_Length"].fillna(penguin["Flipper_Length"].mean(), inplace = True)
penguin.info()
```

```python
# in situtions where you want to remove/discard rows that have null values
# droprows
penguin["Culmen_Depth"].dropna(inplace = True)
penguin.info()
```
