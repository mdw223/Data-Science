## Part 2 Describe variables

```python
import numpy as np
from tabulate import tabulate

D = "Discrete"
O = "Ordinal"
N = "Nominal"
C = "Continuous"


header = ["Column", "Numpy", "Native", "Type of Data"]
arr = []
for col in df.columns:
  dt = df[col].dtype
  native = str(type(df[col][1]))
  obj = ""
  if (col == 'PassengerId'):
    obj = D
  elif (col == 'Survived'):
    obj = N
  elif col == "Pclass":
    obj = O
  elif col == "Name":
    obj = N
  elif col == "Sex":
    obj = N
  elif col == "Age":
    obj = C
  elif col == "SibSp":
    obj = C
  elif col == "Parch":
    obj = D
  elif col == "Ticket":
    obj = N
  elif col == "Fare":
    obj = C
  elif col == "Cabin":
    obj = N
  elif col == "Embarked":
    obj = N

  arr.append([str(col), dt, native, obj])

print(tabulate(arr, headers=header, tablefmt="grid"))
```

###Summary Stats

```python
mean = valid_age_survivor_df['Age'].mean()
median = valid_age_survivor_df['Age'].median()
print("Mean Age: %.2f" % mean)
print("Max Age: %.2d" % valid_age_survivor_df['Age'].max())
print("Min Age: %.2f" % valid_age_survivor_df['Age'].min())
print("Median Age: %.2f" % median)
print("Std Age: %.2f" % valid_age_survivor_df['Age'].std())
print("\nDescription of the distribution of the age of passengers who survived: \n")
print("Shape of Distribution: right skewed")
print("Missing values: No because they were filtered out. But there was a total of %d passengers with missing age values before data cleaning."
      % survivor_df['Age'].isnull().sum())

def find_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
print("Outliers: Passengers with really old age. \nOutliers\n")
outliers = find_outliers_iqr(valid_age_survivor_df, 'Age')
print(outliers)

print("\nThe number of modes: %d" % len(valid_age_survivor_df['Age'].mode()))
print("Distribution: centered around %d as most passengers are around %d years old, with roughly half being around %d years old." %
      (median, mean, int(list(valid_age_survivor_df['Age'].describe())[5])))
print("Spread (Range): %.2f" % (valid_age_survivor_df['Age'].max() - valid_age_survivor_df['Age'].min()))
```
